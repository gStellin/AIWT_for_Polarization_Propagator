# <font color=#DC143C size="+3">Automated generation of Wick's-theorem contributions </font> 
<font color=#9D9D9C size ="+2.25"> at first order in perturbation theory for Gorkov's polarization propagator</font> </br> <font color=#000000 size ="+1">with a single two-body interaction vertex</font>

 ## <font color=#FA8072>Conventions</font>

<p align="justify">From the application of Wick's theorem to the following matrix element (cf. expansion formula of Gorkov's polarization propagator), </p>

$\Pi_{acdb}^{g_1g_3g_4g_2}(t,t')\lvert_{\mathrm{first~order}}\equiv -i \left(\frac{-i}{\hbar}\right) \frac{1}{1!}\int\mathrm{d}t_1 \langle \Phi_0 \lvert T\{\bar{V}(t_1) {A_{I}}_a^{g_1}(t){A_{I}}_b^{g_2}(t'){A_{I}}_d^{\dagger g_4}(t') {A_{I}}_c^{\dagger g_3}(t)\}\lvert\Phi_0\rangle_{\mathrm{conn}}$

the fully-contracted terms consisting of $8$ second-quantization operators are generated. The two-body potential insertion, $\bar{V}(t_1)$, takes the form

$\bar{V}(t_1) = \frac{1}{4} \sum_{pqrs} \bar{v}_{pqrs} {a_I}_p^{\dagger}(t_1){a_I}_q^{\dagger}(t_1) {a_I}_s(t_1) {a_I}_r(t_1)~.$

<p align="justify">Due to the presence of a two-body interaction vertex, the number of distinct fully-contracted terms corresponds to $7!! = 105$. The result is the same for each of the $16$ Nambu components of Gorkov's polarization propagator.  A significant part of these Wick's-theorem contributions has to be discarded, since it corresponds either to disconnected diagrams or to exchange diagrams that are cancelled by the subtractive term of the polarization propagator.</p>

```python
7*5*3*1
```

The alphabetic single-particle indices of the second-quantization operators are converted into positive integers by means of the following map:

<table>
    <tr>

        <th> q </th>
        <th> s </th> 
        <th> r </th>
        <th> a </th> 
        <th> b </th>
        <th> d </th>
        <th> c </th>
    </tr>
    <tr>
        <th> 1 </th> 
        <th> 2 </th>
        <th> 3 </th>
        <th> 4 </th>
        <th> 5 </th> 
        <th> 6 </th>
        <th> 7 </th> 
        <th> 8 </th>
    </tr>
</table>

as such, the single particle indices are stored into arrays (cf. 'Acomb', 'Acombsort', 'AdisjBogoDiag' et c.).

 ## <font color=#FA8072>Generation of the first-order fully-contracted terms</font>

The first-order fully-contracted terms are found to be $105$ in total, consistently with the formula $4n+2m+3!!$ with $n=1$ and $m=0$. Binary contractions belonging to the same fully contracted term (index $i$ in the array initialization) are denoted by the index $j$. The two single particle indices within the same contraction are distinguished by the index $k$. The process produces no repeated pairs, by construction.

```python
l = 1;
Acomb = Table[0, {i, 1, 150}, {j, 1, 4}, {k, 1, 2}];
Acomb[[l]] = {{1, 2}, {3, 4}, {5, 6}, {7, 8}};
For[g = 1, g <= 7, g++,
  growind = 1;
  gcolind = 2;
  For[h = 1, h <= 5, h++,
   hrowind = 2;
   hcolind = 2;
   For[i = 1, i <= 3, i++,
    irowind = 3;
    icolind = 2;
    Acomb[[l]] = Acomb[[1]];
    If[g != 1,
     growpiv = Floor[g/2] + 1;
     gcolpiv = - Floor[g/2] + Ceiling[g/2] + 1;
     Acomb[[l, growind, gcolind]] = Acomb[[1, growpiv, gcolpiv]];
     Acomb[[l, growpiv, gcolpiv]] = Acomb[[1, growind, gcolind]]];
    If[h != 1,
     hrowpiv = Floor[h/2] + 2;
     hcolpiv = - Floor[h/2] + Ceiling[h/2] + 1;
     Atemp1 = Acomb[[l, hrowind, hcolind]];
     Atemp2 = Acomb[[l, hrowpiv, hcolpiv]];
     Acomb[[l, hrowind, hcolind]] = Atemp2;
     Acomb[[l, hrowpiv, hcolpiv]] = Atemp1];
    If[i != 1,
     irowpiv = Floor[i/2] + 3;
     icolpiv = - Floor[i/2] + Ceiling[i/2] + 1;
     Atemp1 = Acomb[[l, irowind, icolind]];
     Atemp2 = Acomb[[l, irowpiv, icolpiv]];
     Acomb[[l, irowind, icolind]] = Atemp2;
     Acomb[[l, irowpiv, icolpiv]] = Atemp1];
    l++]];
  If[g == 7, 
   Print["Loop ", g, " completed and ", l - 1, 
    " Wick's-theorem contributions generated."]]];
```

Sorting code for the generated fully-contracted terms. Single-particle indices within the same contraction are sorted in ascending order.

```python
Acombsort = Table[0, {i, 1, 105}, {j, 1, 4}, {k, 1, 2}];
For[s = 1, s <= 105, s++,
  If[Acomb[[s, 1, 1]] > Acomb[[s, 1, 2]],
   Acombsort[[s, 1, 1]] = Acomb[[s, 1, 2]];
   Acombsort[[s, 1, 2]] = Acomb[[s, 1, 1]],
   Acombsort[[s, 1]] = Acomb[[s, 1]];];
  If[Acomb[[s, 2, 1]] > Acomb[[s, 2, 2]],
   Acombsort[[s, 2, 1]] = Acomb[[s, 2, 2]];
   Acombsort[[s, 2, 2]] = Acomb[[s, 2, 1]],
   Acombsort[[s, 2]] = Acomb[[s, 2]];];
  If[Acomb[[s, 3, 1]] > Acomb[[s, 3, 2]],
   Acombsort[[s, 3, 1]] = Acomb[[s, 3, 2]];
   Acombsort[[s, 3, 2]] = Acomb[[s, 3, 1]],
   Acombsort[[s, 3]] = Acomb[[s, 3]];];
  If[Acomb[[s, 4, 1]] > Acomb[[s, 4, 2]],
   Acombsort[[s, 4, 1]] = Acomb[[s, 4, 2]];
   Acombsort[[s, 4, 2]] = Acomb[[s, 4, 1]],
   Acombsort[[s, 4]] = Acomb[[s, 4]];];];
```

### <font color=#E9967A>Filtration of the fully-contracted terms</font>

Filtration of the conjoint skeleton diagrams. At first order, transpositions between single-particle indices of different vertices (external permutations) are not possible and the overall corresponding combinatorial factor is 1!. On the other hand, transposition between single-particle index of the same two-body vertex are available. Fully-contracted terms which are mapped one another by the permutation $p \mapsto q$  and $r \mapsto s$  and $p \mapsto q \wedge r \mapsto s$ are identified and summed, eventually canceling out the $(2!)^2$ at the denominator, arising from the partial antisymmetry of the $\bar{V}_{pqrs}$ interaction vertex. The cancellation is only partial when one-body propagators linking a vertex to itself are present.

```python
AccVertex = Table[0, {i, 1, 24}, {j, 1, 4}, {k, 1, 2}];
MultccVertex = Table[0, {i, 1, 24}];
Ifrepeat = False;
Temp = Table[0, {i, 1, 4}, {j, 1, 4}, {k, 1, 2}];
IfTemp = Table[0, {i, 1, 4}, {k, 1, 4}];

s = 1;
AccVertex[[1]] = Aconjsort[[1]];
MultccVertex[[1]] = 1;
For[u = 1, u <= 24, u++,
 For[i = 1, i <= 4, i++,
  For[j = 1, j <= 2, j++,
   If[Aconjsort[[u, i, j]] == 1, Temp[[1, i, j]] = 2];
   If[Aconjsort[[u, i, j]] == 2, Temp[[1, i, j]] = 1];
   If[Aconjsort[[u, i, j]] >= 3, 
    Temp[[1, i, j]] = Aconjsort[[u, i, j]]];
   If[Aconjsort[[u, i, j]] == 1, Temp[[2, i, j]] = 2];
   If[Aconjsort[[u, i, j]] == 2, Temp[[2, i, j]] = 1];
   If[Aconjsort[[u, i, j]] == 3, Temp[[2, i, j]] = 4];
   If[Aconjsort[[u, i, j]] == 4, Temp[[2, i, j]] = 3];
   If[Aconjsort[[u, i, j]] >= 5, 
    Temp[[2, i, j]] = Aconjsort[[u, i, j]]];
   If[Aconjsort[[u, i, j]] == 3, Temp[[3, i, j]] = 4];
   If[Aconjsort[[u, i, j]] == 4, Temp[[3, i, j]] = 3];
   If[Aconjsort[[u, i, j]] >= 5 || Aconjsort[[u, i, j]] <=  2, 
    Temp[[3, i, j]] = Aconjsort[[u, i, j]]];
   Temp[[4, i, j]] = Aconjsort[[u, i, j]];
   ]];
 For[i = 1, i <= 4, i++,
  If[Temp[[i, 1, 1]] > Temp[[i, 1, 2]], t2 = Temp[[i, 1, 2]]; 
   t1 = Temp[[i, 1, 1]]; Temp[[i, 1, 1]] = t2; Temp[[i, 1, 2]] = t1];
  If[Temp[[i, 2, 1]] > Temp[[i, 2, 2]], t2 = Temp[[i, 2, 2]]; 
   t1 = Temp[[i, 2, 1]]; Temp[[i, 2, 1]] = t2; Temp[[i, 2, 2]] = t1];
  If[Temp[[i, 3, 1]] > Temp[[i, 3, 2]], t2 = Temp[[i, 3, 2]]; 
   t1 = Temp[[i, 3, 1]]; Temp[[i, 3, 1]] = t2; Temp[[i, 3, 2]] = t1];
  If[Temp[[i, 4, 1]] > Temp[[i, 4, 2]], t2 = Temp[[i, 4, 2]]; 
   t1 = Temp[[i, 4, 1]]; Temp[[i, 4, 1]] = t2; Temp[[i, 4, 2]] = t1];];
 Ifrepeat = False;
 If[u != 1,
  For[v = 1, v <= s, v++,
   Vrepeat = False;
   For[i = 1, i <= 4, i++,
    For[j = 1, j <= 4, j++,
     IfTemp[[i, j]] = 0;
     ]];
   For[i = 1, i <= 4, i++,
    For[j = 1, j <= 4, j++,
      For[k = 1, k <= 4, k++,
        If[AccVertex[[v, i]] == Temp[[j, k]], IfTemp[[j, k]] = 1;];
        ];];];
   If[IfTemp[[1, 1]] + IfTemp[[1, 2]] + IfTemp[[1, 3]] + 
      IfTemp[[1, 4]] == 4, Vrepeat = True;];
   If[IfTemp[[2, 1]] + IfTemp[[2, 2]] + IfTemp[[2, 3]] + 
      IfTemp[[2, 4]] == 4, Vrepeat = True;];
   If[IfTemp[[3, 1]] + IfTemp[[3, 2]] + IfTemp[[3, 3]] + 
      IfTemp[[3, 4]] == 4, Vrepeat = True;];
   If[IfTemp[[4, 1]] + IfTemp[[4, 2]] + IfTemp[[4, 3]] + 
      IfTemp[[4, 4]] == 4, Vrepeat = True;];
   If[Vrepeat == True, Ifrepeat = True; 
    MultccVertex[[v]] = MultccVertex[[v]] + 1;];
   ];
  If[Ifrepeat == False , s++; AccVertex[[s]] = Aconjsort[[u]]; 
   MultccVertex[[s]] = 1;];
  ];
 If[u == 24, 
  Print["Loop ", u, " completed and ", s, 
   " inequivalent conjoint contributions found."]];
 ]
```
