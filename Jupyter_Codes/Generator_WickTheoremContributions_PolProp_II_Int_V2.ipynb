 ## <font color=#FA8072>Conventions</font>

<p align="justify">From the application of Wick's theorem to the following matrix element (cf. expansion formula of Gorkov's polarization propagator), </p>

$\Pi_{acdb}^{g_1g_3g_4g_2}(t,t')\lvert_{\mathrm{second~order}}\equiv -i \left(\frac{-i}{\hbar}\right)^2 \frac{1}{2!}\int\mathrm{d}t_1 \int\mathrm{d}t_2 \langle \Phi_0 \lvert T\{\bar{V}(t_1) \bar{V}(t_2) {A_{I}}_a^{g_1}(t){A_{I}}_b^{g_2}(t'){A_{I}}_d^{\dagger g_4}(t') {A_{I}}_c^{\dagger g_3}(t)\}\lvert\Phi_0\rangle_{\mathrm{conn}}$

the fully-contracted terms consisting of $12$ second-quantization operators are generated. The two-body potential insertions, $\bar{V}(t_1)$ and $\bar{V}(t_2)$, take the form

$\bar{V}(t_1) = \frac{1}{4} \sum_{pqrs} \bar{v}_{pqrs} {a_I}_p^{\dagger}(t_1){a_I}_q^{\dagger}(t_1) {a_I}_s(t_1) {a_I}_r(t_1)~,$

$\bar{V}(t_2) = \frac{1}{4} \sum_{tuvw} \bar{v}_{tuvw} {a_I}_t^{\dagger}(t_2){a_I}_u^{\dagger}(t_2) {a_I}_w(t_2) {a_I}_v(t_2)~.$

Due to the presence of two two-body interaction vertices, the number of distinct fully-contracted terms corresponds to $11!! = 10,395$. The result is the same for each of the $16$ Nambu components of Gorkov's polarization propagator.  A significant part of these Wick's-theorem contributions has to be discarded, since it corresponds either to disconnected diagrams or to exchange diagrams that are cancelled by the subtractive term of the polarization propagator.

 ## <font color=#FA8072>Generation of the first-order fully-contracted terms</font>

The second-order fully-contracted terms are found to be $10,395$ in total, consistently with the formula $4n+2m+3!!$ with $n=2$ and $m=0$. Binary contractions belonging to the same fully contracted term (index $i$ in the array initialization) are denoted by the index $j$. The two single particle indices within the same contraction are distinguished by the index $k$. The process produces no repeated pairs, by construction.

The procedure is repeated $2$ times, assuming in the second case, a permutation of the indices of the two vertices.
