 ## <font color=#FA8072>Conventions</font>

<p align="justify"> The fully-contracted terms processed henceforth correspond to fully-contracted right-dressed disjoint direct contributions generated by the application of Wick's theorem to the following matrix element (cf. expansion formula of Gorkov's polarization propagator), </p>

$\Pi_{acdb}^{g_1g_3g_4g_2}\lvert_{\mathrm{second~order}}\equiv -i \left(\frac{-i}{\hbar}\right)^2 \frac{1}{2!}\int\mathrm{d}t_1 \int\mathrm{d}t_2 \langle \Phi_0 \lvert T\{\bar{V}(t_1)\bar{V}(t_2){A_{I}}_a^{g_1}(t){A_{I}}_b^{g_2}(t'){A_{I}}_d^{\dagger g_4}(t') {A_{I}}_c^{\dagger g_3}(t)\}\lvert\Phi_0\rangle_{\mathrm{conn}}$

where the two-body potential insertions, $\bar{V}(t_1)$ and $\bar{V}(t_2)$, take the form

$\bar{V}(t_1) = \frac{1}{4} \sum_{pqrs} \bar{v}_{pqrs} {a_I}_p^{\dagger}(t_1){a_I}_q^{\dagger}(t_1) {a_I}_s(t_1) {a_I}_r(t_1)~,$

$\bar{V}(t_2) = \frac{1}{4} \sum_{tuvw} \bar{v}_{tuvw} {a_I}_t^{\dagger}(t_2){a_I}_u^{\dagger}(t_2) {a_I}_w(t_2) {a_I}_v(t_2)~.$

Normal (anomalous) components possess an even (even or odd) number of '1' and '2' indices. ### <font color=#E9967A>Definition of the auxiliary arrays</font>

<p align="justify">Arrays specifying the Nambu indices (cf. 'NambuDirRight'), the operator type (cf.'SQTypeDirRight') and distinguishing internal from external single-particle indices (cf. 'IntExtDirRight') are defined, in accordance with the conventions for the second-quantization operators in the header of the notebook. In particular, to external (internal) single-particle indices the value of '1' ('0') is assigned, whereas for second-quantization operators of creation (annihilation) type the associated array entry is '0' ('1'). For Nambu indices, the convention is straightforward.</p>

```python
NambuDirRight = Table[0, {i, 1, 38}, {j, 1, 6}, {k, 1, 2}];
For[r = 1, r <= 38, r++, 
   For[j = 1, j <= 6, j++, 
     For[k = 1, k <= 2, k++, 
       If[CombDirRight[[r, j, k]] <= 8, NambuDirRight[[r, j, k]] = 1];
       If[CombDirRight[[r, j, k]] == 9, NambuDirRight[[r, j, k]] = GPPNambu[[NppIdx, 1]]]; 
       If[CombDirRight[[r, j, k]] == 10, NambuDirRight[[r, j, k]] = GPPNambu[[NppIdx, 4]]]; 
       If[CombDirRight[[r, j, k]] == 11, NambuDirRight[[r, j, k]] = GPPNambu[[NppIdx, 3]]]; 
       If[CombDirRight[[r, j, k]] == 12, NambuDirRight[[r, j, k]] = GPPNambu[[NppIdx, 2]]]; 
      ]; 
    ]; 
  ];
```

### <font color=#E9967A>Initialization of Gorkov-Green's functions</font>

An anomalous (or '2') Nambu index is introduced whenever a homogeneous pair of second-quantization operators is found, consistently with the definition of the one-body Gorkov propagators (cf. Somà et al, 2011). In Bogoliubov diagrams, involute single-particle indices do appear, since at least an anomalous contraction of external single-particle indices is present. In this case, involute single-particle indices are mapped into negative natural numbers. 

```python
GGFNambu = Table[0, {i, 1, 38}, {j, 1, 6}, {k, 1, 2}];
GGFTime = Table[0, {i, 1, 38}, {j, 1, 6}, {k, 1, 2}];
GGFSParticle = Table[0, {i, 1, 38}, {j, 1, 6}, {k, 1, 2}]; 
```

```python
For[r = 1, r <= 38, r++, 
   For[i = 1, i <= 6, i++, 
    (* Canonical contractions *) 
    (* Contraction between a non-involute annihilation operator
    followed by an involute annihilation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 2 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
        GGFNambu[[r, i, 1]] = 1; 
        GGFNambu[[r, i, 2]] = 2; 
       ]; 
     ]; 
    (* Contraction between an involute annihilation operator
    followed by a non-involute annihilation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 2]] == 1 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
        GGFNambu[[r, i, 1]] = 1; 
        GGFNambu[[r, i, 2]] = 2; 
       ]; 
     ]; 
    (* Contraction between a non-involute creation operator
    followed by an involute creation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 2 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
        GGFNambu[[r, i, 1]] = 2; 
        GGFNambu[[r, i, 2]] = 1; 
       ]; 
     ]; 
    (* Contraction between an involute creation operator
    followed by a non-involute creation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 2]] == 1 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
        GGFNambu[[r, i, 1]] = 2; 
        GGFNambu[[r, i, 2]] = 1; 
       ]; 
     ]; 
    (* Contraction between a non-involute annihilation operator
    followed by a non-involute creation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 1 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
        GGFNambu[[r, i, 1]] = 1; 
        GGFNambu[[r, i, 2]] = 1; 
       ]; 
     ]; 
    (* Contraction between an involute annihilation operator
    followed by an involute creation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 2]] == 2 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
        GGFNambu[[r, i, 1]] = 2; 
        GGFNambu[[r, i, 2]] = 2; 
       ]; 
     ]; 
    (* Contraction between a non-involute creation operator
    followed by a non-involute annihilation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 1 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
        GGFNambu[[r, i, 1]] = 1; 
        GGFNambu[[r, i, 2]] = 1; 
       ]; 
     ]; 
    (* Contraction between an involute creation operator
    followed by an involute annihilation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 2]] == 2 , 
        GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
        GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
        GGFNambu[[r, i, 1]] = 2; 
        GGFNambu[[r, i, 2]] = 2; 
       ]; 
     ]; 
    (* Non-canonical contractions *) 
    (* Contractions with two non-involute annihilation operators *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 1 , 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
       ]; 
     ]; 
    (* Contractions with two involute annihilation operators *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 1]] == 2, 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
       ]; 
     ]; 
    (* Contractions with two non-involute creation operators *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 1 , 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
       ]; 
     ]; 
    (* Contractions with two involute creation operators *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 1]] == 2, 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
       ]; 
     ]; 
    (* Contraction between a non-involute annihilation operator 
    followed by an involute creation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 2 , 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
       ]; 
     ]; 
    (* Contraction between a involute annihilation operator 
    followed by an non-involute creation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 1 && SQTypeDirRight[[r, i, 2]] == 0, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 2]] == 1 , 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
       ]; 
     ]; 
    (* Contraction between a involute creation operator 
    followed by an non-involute annihilation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 2 && NambuDirRight[[r, i, 2]] == 1 , 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
       ]; 
     ]; 
    (* Contraction between a non-involute creation operator 
    followed by an involute annihilation operator *) 
     If[SQTypeDirRight[[r, i, 1]] == 0 && SQTypeDirRight[[r, i, 2]] == 1, 
      If[NambuDirRight[[r, i, 1]] == 1 && NambuDirRight[[r, i, 2]] == 2 , 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 1 && IntExtDirRight[[r, i, 2]] == 0, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 2]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 1]]; 
         GGFNambu[[r, i, 1]] = 1; 
         GGFNambu[[r, i, 2]] = 1; 
        ]; 
        If[IntExtDirRight[[r, i, 1]] == 0 && IntExtDirRight[[r, i, 2]] == 1, 
         GGFSParticle[[r, i, 1]] = CombDirRight[[r, i, 1]]; 
         GGFSParticle[[r, i, 2]] = CombDirRight[[r, i, 2]]; 
         GGFNambu[[r, i, 1]] = 2; 
         GGFNambu[[r, i, 2]] = 2; 
        ]; 
       ]; 
     ]; 
    ];];
```

### <font color=#E9967A>Determination of the sign</font>

The sign corresponds to the minimum number of 2-cycle/binary permutations (<i>i.e.</i> transpositions) necessary to disentangle the all the contractions of second-quantization operators and, thus, reproduce the loaded sequences of single-particle indices.

```python
AmplSign = Table[0, {i, 1, 38}];
SPSequence = Table[0, {i, 1, 38}, {j, 1, 12}];
```

```python
For[r = 1, r <= 38, r++, 
   For[i = 0, i <= 5, i++, 
    SPSequence[[r, 2 i + 1]] = Abs[GGFSParticle[[r, i + 1, 1]]]; 
    SPSequence[[r, 2 i + 2]] = Abs[GGFSParticle[[r, i + 1, 2]]]; 
   ]];
```

```python
For[r = 1, r <= 38, r++, 
   c = 0; 
   For[i = 1, i <= 12, i++, 
    For[j = i + 1, j <= 12, j++, 
     If[SPSequence[[r, i]] > SPSequence[[r, j]], c = c + 1];]]; 
   AmplSign[[r]] = Mod[c, 2]; 
  ];
```

### <font color=#E9967A>Assignment of the time indices, based on the canonical ordering</font>

<html>Times are sorted in ascending order: '4' is the latest time and denotes t, '3' denotes t<sub>1</sub>, '2' denotes t<sub>2</sub>, whereas the earliest time is t' and is denoted by '1'.</html>

```python
For[r = 1, r <= 38, r++, 
  For[i = 1, i <= 6, i++,
   For[j = 1, j <= 2, j++,
    If[GGFSParticle[[r, i, j]] == 1 || GGFSParticle[[r, i, j]] == 2 || 
      GGFSParticle[[r, i, j]] == 3 ||  GGFSParticle[[r, i, j]] == 4 , 
     GGFTime[[r, i, j]] = 3]; (* Time Class 3 = Subscript[t, 1] *)
    If[GGFSParticle[[r, i, j]] == 5 || GGFSParticle[[r, i, j]] == 6 || 
      GGFSParticle[[r, i, j]] == 7 ||  GGFSParticle[[r, i, j]] == 8 , 
     GGFTime[[r, i, j]] = 2]; (* Time Class 2 = Subscript[t, 2] *)
    If[GGFSParticle[[r, i, j]] == 10 || GGFSParticle[[r, i, j]] == 11,
      GGFTime[[r, i, j]] = 1]; (* Time Class 1 = t' *)
    If[GGFSParticle[[r, i, j]] == 9 || GGFSParticle[[r, i, j]] == 12, 
     GGFTime[[r, i, j]] = 4];(*  Time Class 4 = t *)
    ]]];
```

### <font color=#E9967A>Assignment of the time indices, based on the time ordering</font>

It is convenient to assign the times (associated with the single-particle and the Nambu indices) only once the time ordering is established. This permits to evaluate the Gorkov-Green's functions more rapidly, exploiting the order of the natural numbers 1, 2, 3 and 4. The time indices appearing henceforth are, thus, relative times (= time-ordering dependent), whereas the absolute times (cf. the convention in the opening) correspond to the time classes. Times are sorted in ascending order: '4' is the latest time and '1' is the earliest one. 

```python
T = 1;
For[r = 1, r <= 38, r++, 
  For[i = 1, i <= 6, i++,
   For[j = 1, j <= 2, j++,
    If[GGFSParticle[[r, i, j]] == 1 || GGFSParticle[[r, i, j]] == 2 || 
      GGFSParticle[[r, i, j]] == 3 ||  GGFSParticle[[r, i, j]] == 4 , 
     GGFTime[[r, i, j]] = TimeClassOrdPlus[[T, 3]]]; (* Time Class 3 = t_1 *)
    If[GGFSParticle[[r, i, j]] == 5 || GGFSParticle[[r, i, j]] == 6 || 
      GGFSParticle[[r, i, j]] == 7 ||  GGFSParticle[[r, i, j]] == 8 , 
     GGFTime[[r, i, j]] = TimeClassOrdPlus[[T, 2]]]; (* Time Class 2 = t_2 *)
    If[GGFSParticle[[r, i, j]] == 10 || GGFSParticle[[r, i, j]] == 11,
      GGFTime[[r, i, j]] = TimeClassOrdPlus[[T, 1]]]; (* Time Class 1 = t' *)
    If[GGFSParticle[[r, i, j]] == 9 || GGFSParticle[[r, i, j]] == 12, 
     GGFTime[[r, i, j]] = TimeClassOrdPlus[[T, 4]]](*  Time Class 4 = t *)
    ]]];  
```

```python
Clear[k];
GGFAmplitude = Table[0, {i, 1, 38}, {j, 1, 6}];
GGFPolynomial = Table[0.0, {i, 1, 38}, {j, 1, 6}];

For[r = 1, r <= 38, r++, 
  For[i = 1, i <= 6, i++,
   If[GGFTime[[r, i, 1]] > GGFTime[[r, i, 2]], 
    GGFAmplitude[[r, i]] = StringForm["``", Superscript[
        Superscript[Subscript[𝜒, NtL[GGFSParticle[[r, i, 1]]]], 
         GGFNambu[[r, i, 1]]], Subscript[k, i]]] StringForm["``", 
       Superscript[Superscript[Subscript[OverBar[𝜒], NtL[GGFSParticle[[r, i, 2]]]], 
         GGFNambu[[r, i, 2]]], Subscript[k, i]]];
    GGFPolynomial[[r,i]] = -(Subscript[Ω, Subscript[k, i]] - 
          Subscript[Ω, 0]) Subscript[t,GGFTime[[r, i, 1]]]/ℏ + ( 
        Subscript[Ω, Subscript[k, i]] - Subscript[Ω, 0]) Subscript[t,GGFTime[[r, i, 2]]]/ℏ;];
   If[GGFTime[[r, i, 1]] < GGFTime[[r, i, 2]], GGFAmplitude[[r,i]] = -StringForm["``", 
        Superscript[Superscript[Subscript[Υ, NtL[GGFSParticle[[r, i, 1]]]], 
          GGFNambu[[r, i, 1]]], Subscript[k, i]]] StringForm["``", Superscript[
        Superscript[Subscript[OverBar[Υ],NtL[GGFSParticle[[r, i, 2]]]], GGFNambu[[r, i, 2]]], 
        Subscript[k, i]]];
     GGFPolynomial[[r, i]] = +(Subscript[Ω, Subscript[k, i]] - Subscript[Ω, 0]) Subscript[t, 
        GGFTime[[r, i, 1]]]/ℏ - (Subscript[Ω, Subscript[k, i]] - Subscript[Ω, 0]) Subscript[t, 
        GGFTime[[r, i, 2]]]/ℏ;];
   If[GGFTime[[r, i, 1]] == GGFTime[[r, i, 2]], 
    If[GGFNambu[[r, i, 1]] == 2  &&  GGFNambu[[r, i, 2]] == 2, 
     GGFAmplitude[[r, i]] = -StringForm["``",Superscript[
          Superscript[Subscript[Υ, NtL[GGFSParticle[[r, i, 1]]]],
            GGFNambu[[r, i, 1]]], Subscript[k, i]]] StringForm["``", 
        Superscript[Superscript[Subscript[OverBar[Υ],NtL[GGFSParticle[[r, i, 2]]]], GGFNambu[[r, i, 2]]], 
         Subscript[k, i]]];
      GGFPolynomial[[r,i]] = +(Subscript[Ω, Subscript[k, i]] - Subscript[Ω, 0]) Subscript[t, 
         GGFTime[[r, i, 1]]]/ℏ- (Subscript[Ω, Subscript[k, i]] - Subscript[Ω, 0]) Subscript[t,
         GGFTime[[r, i, 2]]]/ℏ, 
     GGFAmplitude[[r, i]] = StringForm["``", Superscript[
         Superscript[Subscript[𝜒, NtL[GGFSParticle[[r, i, 1]]]], 
          GGFNambu[[r, i, 1]]], Subscript[k, i]]] StringForm["``", Superscript[
         Superscript[Subscript[OverBar[𝜒], NtL[GGFSParticle[[r, i, 2]]]], 
          GGFNambu[[r, i, 2]]], Subscript[k, i]]];
     GGFPolynomial[[r, i]] = -(Subscript[Ω, Subscript[k, i]] - Subscript[Ω, 0]) Subscript[t, 
         GGFTime[[r, i, 1]]]/ℏ + (Subscript[Ω, Subscript[k, i]] - Subscript[Ω, 0]) Subscript[t, 
         GGFTime[[r, i, 2]]]/ℏ]];
   ]];
```

### <font color=#E9967A>Calculation of the amplitude associated to Goldstone diagrams with alphabetic indices</font>

<html>The three time integrals are evaluated in two steps. First, one computes the integrals over t<sub>1</sub> and t<sub>2</sub> (dummy times), by solving a linear system with a parameter (the third frequency). At second order, the frequency variables which are evaluated after integration by $t_1$ and $t_2$ is labeled as $\gamma$ and $\epsilon$. Consequently, the time difference multiplying $\gamma$ and $\epsilon$ must contain the times $t_1$ and $t_2$, otherwise the integration fails. Next, the integration over (t-t') is performed by solving a single linear equation, obtained by equating the coefficient of (t-t') in the total polynomial to zero. Note that the time interval that multiplies &#989; must contain either t or t'. However, t and t' shall not appear together in that time interval, otherwise the time ordering of the corresponding Goldstone diagram becomes undefined (or ambiguously defined). Consequently, in the denominators of the energy representation, two $\omega$'s
may appear, depending on the topology of the unoriented Goldstone diagram. Besides, the time interval multiplying &#989; shall never coincide with the one associated with $\gamma$ or $\epsilon$, otherwise the implemented time sequence is ill-defined.</html>

```python
GGFTotalPolynomial = Table[0, {r, 1, 38}];

For[r = 1, r <= 38, r++,
  For[j = 1, j <= 6, j++,
   GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] + GGFPolynomial[[r, j]];
   ];
   If[T == 1, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,3] - Subscript[t,2])];
   If[T == 2, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 3, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 4, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 5, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 6, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];  
   If[T == 7, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];   
   If[T == 8, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,3] - Subscript[t,2])];  
   If[T == 9, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,3] - Subscript[t,2]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,4] - Subscript[t,3])];
   If[T == 10, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,2] - Subscript[t,1]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,4] - Subscript[t,3])];
   If[T == 11, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,3] - Subscript[t,2])];
   If[T == 12, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdPlus[[T, 4]]] - Subscript[t,TimeClassOrdPlus[[T, 1]]])
   - 𝛾 (Subscript[t,2] - Subscript[t,1]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,4] - Subscript[t,3])];
  ];
```

### <font color=#E9967A>Calculation of the amplitude associated to Goldstone diagrams with alphabetic indices</font>

<html>The three time integrals are evaluated in two steps. First, one computes the integrals over t<sub>1</sub> and t<sub>2</sub> (dummy times), by solving a linear system with a parameter (the third frequency). At second order, the frequency variables which are evaluated after integration by $t_1$ and $t_2$ is labeled as $\gamma$ and $\epsilon$. Consequently, the time difference multiplying $\gamma$ and $\epsilon$ must contain the times $t_1$ and $t_2$, otherwise the integration fails. Next, the integration over (t-t') is performed by solving a single linear equation, obtained by equating the coefficient of (t-t') in the total polynomial to zero. Note that the time interval that multiplies &#989; must contain either t or t'. However, t and t' shall not appear together in that time interval, otherwise the time ordering of the corresponding Goldstone diagram becomes undefined (or ambiguously defined). Consequently, in the denominators of the energy representation, two $\omega$'s
may appear, depending on the topology of the unoriented Goldstone diagram. Besides, the time interval multiplying &#989; shall never coincide with the one associated with $\gamma$ or $\epsilon$, otherwise the implemented time sequence is ill-defined.</html>

```python
GGFTotalPolynomial = Table[0, {r, 1, 38}];

For[r = 1, r <= 38, r++,
  For[j = 1, j <= 6, j++,
   GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] + GGFPolynomial[[r, j]];
   ];
   If[T == 1, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,3] - Subscript[t,2]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,4] - Subscript[t,3])];
   If[T == 2, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,3] - Subscript[t,2])];
   If[T == 3, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 4, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,2] - Subscript[t,1]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,4] - Subscript[t,3])];
   If[T == 5, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 6, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,3] - Subscript[t,2])];  
   If[T == 7, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];   
   If[T == 8, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,3] - Subscript[t,2])];  
   If[T == 9, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 10, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 11, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,3] - Subscript[t,2]) 
   - ϝ (Subscript[t,2] - Subscript[t,1])];
   If[T == 12, GGFTotalPolynomial[[r]] = GGFTotalPolynomial[[r]] 
   + 𝜔 (Subscript[t, TimeClassOrdMinus[[T, 4]]] - Subscript[t,TimeClassOrdMinus[[T, 1]]])
   - 𝛾 (Subscript[t,4] - Subscript[t,3]) - 𝜖 (Subscript[t,2] - Subscript[t,1]) 
   - ϝ (Subscript[t,3] - Subscript[t,2])];
  ];
```
