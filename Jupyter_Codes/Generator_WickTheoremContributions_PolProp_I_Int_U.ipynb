 ## <font color=#FA8072>Conventions</font>

<p align="justify">From the application of Wick's theorem to the following matrix element (cf. expansion formula of Gorkov's polarization propagator), </p>

$\Pi_{acdb}^{g_1g_3g_4g_2}(t,t')\lvert_{\mathrm{first~order}}\equiv +i \left(\frac{-i}{\hbar}\right) \frac{1}{1!}\int\mathrm{d}t_1 \langle \Phi_0 \lvert T\{U(t_1) {A_{I}}_a^{g_1}(t){A_{I}}_b^{g_2}(t'){A_{I}}_d^{\dagger g_4}(t') {A_{I}}_c^{\dagger g_3}(t)\}\lvert\Phi_0\rangle_{\mathrm{conn}}$

the fully-contracted terms consisting of 6 second-quantization operators are generated. The one-body potential insertion, $U(t_1)$, takes the form

$U(t_1) = \frac{1}{2}\sum_{ef}[u_{ef}^{11}{a_I}_e^{\dagger}(t_1){a_I}_f(t_1)+u_{ef}^{22}{a_I}_{\bar{e}}(t_1){a_I}_{\bar{f}}^{\dagger}(t_1)+u_{ef}^{12}{a_I}_e^{\dagger}(t_1){a_I}_{\bar{f}}^{\dagger}(t_1)+u_{ef}^{21}{a_I}_{\bar{e}}(t_1){a_I}_f(t_1)]~,$

<p align="justify">Due to the presence of a one-body interaction vertex, the number of distinct fully-contracted terms corresponds to $5!! = 15$. The results of the generation process hold identically for all the 4 Nambu components of the one-body potential and are the same for each of the $16$ Nambu components of Gorkov's polarization propagator. A significant part of these Wick's-theorem contributions has to be discarded, since it corresponds either to disconnected diagrams or to exchange diagrams that are cancelled by the subtractive term of the polarization propagator. At any order, no conjoint diagram containing only one-body interaction vertices exists. Additionally, since the one-body potential is not invariant under permutation of the two indices, the filtration process does not apply at first order, as only one interaction vertex is present. The fully-contracted terms are, thus, inequivalent.  </p>

 ## <font color=#FA8072>Generation of the first-order fully-contracted terms</font>

The first-order fully-contracted terms are found to be $15$ in total, consistently with the formula $4n+2m+3!!$ with $n=0$ and $m=1$. Binary contractions belonging to the same fully contracted term (index $i$ in the array initialization) are denoted by the index $j$. The two single particle indices within the same contraction are distinguished by the index $k$. The process produces no repeated pairs, by construction.

```python
l = 1;
Acomb = Table[0, {i, 1, 20}, {j, 1, 3}, {k, 1, 2}];
Acomb[[l]] = {{1, 2}, {3, 4}, {5, 6}};
For[g = 1, g <= 5, g++,
  growind = 1;
  gcolind = 2;
  For[h = 1, h <= 3, h++,
   hrowind = 2;
   hcolind = 2;
   Acomb[[l]] = Acomb[[1]];
   If[g != 1,
    growpiv = Floor[g/2] + 1;
    gcolpiv = - Floor[g/2] + Ceiling[g/2] + 1;
    Acomb[[l, growind, gcolind]] = Acomb[[1, growpiv, gcolpiv]];
    Acomb[[l, growpiv, gcolpiv]] = Acomb[[1, growind, gcolind]]];
   If[h != 1,
    hrowpiv = Floor[h/2] + 2;
    hcolpiv = - Floor[h/2] + Ceiling[h/2] + 1;
    Atemp1 = Acomb[[l, hrowind, hcolind]];
    Atemp2 = Acomb[[l, hrowpiv, hcolpiv]];
    Acomb[[l, hrowind, hcolind]] = Atemp2;
    Acomb[[l, hrowpiv, hcolpiv]] = Atemp1];
   l++];
  If[g == 5, 
   Print["Loop ", g, " completed and ", l - 1, 
    " Wick's-theorem contributions generated."]]];
```

Sorting code for the generated fully-contracted terms. Single-particle indices within the same contraction are sorted in ascending order.

```python
Acombsort = Table[0, {i, 1, 15}, {j, 1, 3}, {k, 1, 2}];
For[s = 1, s <= 15, s++,
  If[Acomb[[s, 1, 1]] > Acomb[[s, 1, 2]],
   Acombsort[[s, 1, 1]] = Acomb[[s, 1, 2]];
   Acombsort[[s, 1, 2]] = Acomb[[s, 1, 1]],
   Acombsort[[s, 1]] = Acomb[[s, 1]];];
  If[Acomb[[s, 2, 1]] > Acomb[[s, 2, 2]],
   Acombsort[[s, 2, 1]] = Acomb[[s, 2, 2]];
   Acombsort[[s, 2, 2]] = Acomb[[s, 2, 1]],
   Acombsort[[s, 2]] = Acomb[[s, 2]];];
  If[Acomb[[s, 3, 1]] > Acomb[[s, 3, 2]],
   Acombsort[[s, 3, 1]] = Acomb[[s, 3, 2]];
   Acombsort[[s, 3, 2]] = Acomb[[s, 3, 1]],
   Acombsort[[s, 3]] = Acomb[[s, 3]];];];
```

# <font color=#71797E>Disjoint Diagrams</font>

Disjoint contributions for each Nambu component of the one-body potential at first order are $12$ in total. <br>
These are distributed as follows: <br>
i) $2$ Bogoliubov diagrams dressed in the antidiagonal; <br>
ii) $2$ Bogoliubov diagrams dressed in the diagonal; <br>
iii) $2$ Exchange diagrams dressed in the propagator above; <br>
iv) $2$ Exchange diagrams dressed in the propagator below; <br>
v) $2$ Direct diagrams dressed in the left propagator; <br>
vi) $2$ Direct diagrams dressed in the right propagator; <br>

Total number of disjoint diagrams:

```python
15 - 3
```

```python
AdisjBogoDiag = Table[0, {i, 1, 12}, {j, 1, 3}, {k, 1, 2}];
AdisjBogoAntidiag = Table[0, {i, 1, 12}, {j, 1, 3}, {k, 1, 2}];
AdisjDirLeft = Table[0, {i, 1, 12}, {j, 1, 3}, {k, 1, 2}];
AdisjDirRight = Table[0, {i, 1, 12}, {j, 1, 3}, {k, 1, 2}];
AdisjExchAbove = Table[0, {i, 1, 12}, {j, 1, 3}, {k, 1, 2}];
AdisjExchBelow = Table[0, {i, 1, 12}, {j, 1, 3}, {k, 1, 2}];
m = 1;
n = 1;
p = 1;
q = 1;
r = 1;
s = 1;
For[i = 1, i <= 15, i++, 
 Bridge34 = False;
 Bridge35 = False;
 Bridge36 = False;
 Bridge45 = False;
 Bridge46 = False;
 Bridge56 = False;
 For[j = 1, j <= 3, j++,
   If[Acombsort[[i, j]] == {3, 4}, Bridge34 = True];
  If[Acombsort[[i, j]] == {3, 5}, Bridge35 = True];
  If[Acombsort[[i, j]] == {3, 6}, Bridge36 = True];
  If[Acombsort[[i, j]] == {4, 5}, Bridge45 = True];
  If[Acombsort[[i, j]] == {4, 6}, Bridge46 = True];
  If[Acombsort[[i, j]] == {5, 6}, Bridge56 = True];
  ];
 If[Bridge35 == True && Bridge46 == False, 
  AdisjDirRight[[p]] = Acombsort[[i]]; p++];
 If[Bridge35 == False && Bridge46 == True, 
  AdisjDirLeft[[q]] = Acombsort[[i]]; q++];
 If[Bridge34 == True && Bridge56 == False, 
  AdisjBogoAntidiag[[m]] = Acombsort[[i]]; m++];
 If[Bridge34 == False && Bridge56 == True, 
  AdisjBogoDiag[[n]] = Acombsort[[i]]; n++];
 If[Bridge36 == False && Bridge45 == True, 
  AdisjExchAbove[[r]] = Acombsort[[i]]; r++];
 If[Bridge36 == True && Bridge45 == False, 
  AdisjExchBelow[[s]] = Acombsort[[i]]; s++];
 If[i == 15, 
  Print["Loop ", i, " completed and ", m - 1, 
   " antidiagonally-dressed Bogoliubov contributions found."] ];
 If[i == 15, 
  Print["Loop ", i, " completed and ", n - 1, 
   " diagonally-dressed Bogoliubov contributions found."] ];
 If[i == 15, 
  Print["Loop ", i, " completed and ", r - 1, 
   " above-dressed exchange contributions found."] ];
 If[i == 15, 
  Print["Loop ", i, " completed and ", s - 1, 
   " below-dressed exchange contributions found."] ];
 If[i == 15, 
  Print["Loop ", i, " completed and ", p - 1, 
   " left-dressed direct contributions found."] ];
 If[i == 15, 
  Print["Loop ", i, " completed and ", q - 1, 
   " right-dressed direct contributions found."] ];]
```
